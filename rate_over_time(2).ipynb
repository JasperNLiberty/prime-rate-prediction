```python
%matplotlib inline

from fbprophet import Prophet

import pandas as pd

import datetime

df = pd.read_csv('./boc.csv')
df = df[::-1]
df = df.dropna()

df = df.rename(columns={'date': 'ds', 'prime': 'y'}).reset_index().drop(['index'], axis=1)

# filter time
df = df[df.ds > '2008-06-17']

df['ds'] = df.ds.apply(lambda x:datetime.datetime.strptime(x, "%Y-%m-%d"))

df['y'] = df.y.apply(lambda x:float(x))

df.plot(x='ds', y='y')

m = Prophet()

m.fit(df)

future = m.make_future_dataframe(periods=365*3+120)
future.tail()

# Python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()

# Python
fig1 = m.plot(forecast)
```
